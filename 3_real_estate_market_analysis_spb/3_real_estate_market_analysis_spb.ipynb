### Дубликаты

В датафрейме есть колонка `locality_name` со строковыми значениями — названиями населённых пунктов. Можно предположить, что здесь возможно наличие неявных дубликатов.
Выведем на экран список уникальных названий населённых пунктов, отсортированный в алфавитном порядке.

```python
df['locality_name'].sort_values().unique()
```

В выведенном списке повторяются аналогичные названия населённых пунктов.

Можно было бы произвести замену буквы `ё` на `е`, чтобы варианты написания типов поселения (`посёлок` и `поселок`) не множили уникальные значения столбца, но ~~я очень люблю букву "ё"~~ в списке также повторяются значения с и без указания типа населённого пункта, например `Кудрово` и `деревня Кудрово` или `Мурино` и `посёлок Мурино`.

Поэтому лучше напишем функцию, которая отбросит из строк в этом столбце все слова, не начинающиеся с заглавной буквы, т.е. сохранит лишь сами названия населённых пунктов без указания их типов.

```python
def drop_locality(row):
    
    locality_name = row['locality_name']
    name_list = []                                  # пустой список для отобранных слов
    
    if pd.isna(row['locality_name']):               # если в строке пропуск,
        return                                      # возвращаем пропуск
    
    else:
        for word in locality_name.split(sep=' '):   # делим строку на отдельные слова по пробелу, и каждое слово
            if word.istitle():                      # проверяем на заглавную букву в его начале
                name_list.append(word)              # добавляем в пустой список все слова, начинающиеся с заглавной буквы
        return ' '.join(name_list)                  # превращаем список обратно в строку и возвращаем новую строку
```

И применим эту функцию к столбцу `locality_name`, сохранив новые укороченные названия в новом столбце `short_locality`.

```python
df['short_locality'] = df.apply(drop_locality, axis=1)

df['short_locality'].sort_values().unique()      # проверка укороченных названий в новом столбце
```

### Обработка пропущенных значений

Теперь перейдём к работе над пропусками значений в датафрейме.

Из ТЗ известно, что часть значений вписаны в таблицу пользователем, другие — получены автоматически на основе картографических данных. Соответственно, заранее можно предположить, что на появление пропусков в первых повлиял человеческий фактор, а во вторых -- ошибки при выгрузке данных.

Пройдёмся по колонкам с пропущенными значениями последовательно. 
Но прежде взглянем на долю пропусков в каждой колонке датафрейма:

```python
df.isna().mean()
```

Многие столбцы имеют довольно высокую долю пропущенных значений. Например, в колонках `airports_nearest` и `city_сenters_nearest` доля пропусков -- примерно по 23 % (а в `parks_nearest` доходит почти до 66 % !). При этом вероятно, количество этих пропусков можно свести вплоть до 0,2 % -- столько пропусков содержится в колонке `locality_name`, по которой можно было бы высчитать расстояния от населённого пункта *до центра Санкт-Петербурга* и до единственного крупного открытого аэропорта в области, на который имеет смысл ссылаться в данных -- *Пулково*.

Однако задача выглядит нетривиальной, что наводит на мысль о целесообразности заполнения всех пропусков.

Изучив поставленные в ТЗ задачи, можно заключить, что вовсе не все столбцы датафрейма необходимы для исследования, а значит на данном этапе можно просто отбросить те из них, на которых не будут основываться параметры.

[Если в ходе исследования я изменю мнение относительно необходимых столбцов, я вернусь к этому пункту]

```python
df.drop(['airports_nearest', 'parks_around3000', 'parks_nearest',
         'ponds_around3000', 'ponds_nearest', 'total_images', 'locality_name'], axis='columns', inplace=True)

df.head()
```

Далее по порядку.

Пропуски в столбце `ceiling_height` резонно заменить медианным значением по стоблцу, поскольку высота потолков зависит от конкретного дома, и проверить эти значения нет возможности.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['ceiling_height'].isna().sum()
```

Переходим к столбцу `floors_total`. Тут всего 0,3 % пропусков, в то же время эти данные важны для рассчёта параметров, а восстановить их также нет возможности. Поэтому немногочисленные строчки с этими пропусками из датафрейма удалим.

```python
df = df.dropna(subset=['floors_total'])
df['floors_total'].isna().sum()
```

Пропусков в колонке `living_area` относительно немного — около 8 %. Можно попробовать заполнить их медианными значениями по столбцу с поправкой на количество комнат в квартирах (т.к. с столбце `rooms` пропусков нет).


Посмотрим, как распределяется "комнатность" по квартирам:

```python
df['rooms'].value_counts()
```

Замену произведём следующим образом:
* квартирам с 0 и 1 комнатой присвоим медианное значение для _rooms <= 1_;
* квартирам с 2мя комнатами — медиану для _rooms = 2_;
* квартирам с 3мя комнатами — медиану для _rooms = 3_;
* остальным квартирам с 4+ комнатами — медиану для _rooms >= 4_.

```python
# напишем функцию для категоризации квартир по количеству комнат
# функция принимает в качестве аргумента количество комнат и возвращает категорию квартиры

def room_category(room):
    if room <= 1:
        return 'A'
    if room == 2:
        return 'B'
    if room == 3:
        return 'C'
    return 'D'

# проверка работы функции:
display(room_category(0))
display(room_category(1))
display(room_category(2))
display(room_category(3))
display(room_category(6))
display(room_category(15))
```

```python
# создадим в таблице новый столбец, запишем в него категории квартир, применив к столбцу 'rooms' новую функцию
df['room_category'] = df['rooms'].apply(room_category)

# заполним пропуски в столбце 'living_area' медианами площадей, используя группировку по новому столбцу и лямбда-функцию
df['living_area'] = df.groupby('room_category')['living_area'].apply(lambda x: x.fillna(x.median()))

# проверяем количество пропусков
display(df['living_area'].isna().sum())

# удаляем столбец с категориями за ненадобностью
df = df.drop(['room_category'], axis='columns')
```

Взглянем на значения колонки `is_apartment`.

```python
df['is_apartment'].unique()
```

Логично предположить, что если человек, заполняющий данные о квартире, не указал ничего в графе "Апартаменты", то квартира не является апартаментами (тем более, что таких пропусков 88 %), а значит эти пропуски можно заменить на значение `False`.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].isna().sum()
```

Площадь кухни зависит скорее от планировки квартиры, поэтому пропуски в `kitchen_area` просто заменим медианой по столбцу.

```python
df['kitchen_area'] = df['kitchen_area'].fillna(df['kitchen_area'].median())
df['kitchen_area'].isna().sum()
```

Аналогично со столбцом про апартаменты предполагаем, что если количество балконов в квартире не было указано, то их нет совсем: заменяем пропуски в столбце `balcony` на нули.

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'].isna().sum()
```

В столбце `short_locality` 0,2 % пропусков: мало, заполнить сложно, данные для исследования нужны — поэтому эти строки тоже удаляем.

```python
df = df.dropna(subset=['short_locality'])
df['short_locality'].isna().sum()
```

Переходим к `city_сenters_nearest`. Здесь пропусков много, поэтому удалить эти строки нельзя. Для анализа они нужны. Попробуем аналогично с жилой площадью квартир заполнить пропущенные значения медианными значениями по населённым пунктам. 

```python
display(f'количество пропусков в столбце "city_сenters_nearest" до заполнения — {df["city_сenters_nearest"].isna().sum()}')

df['city_сenters_nearest'] = df.groupby('short_locality')['city_сenters_nearest'].apply(lambda x: x.fillna(x.median()))

display(f'количество пропусков в столбце "city_сenters_nearest" после заполнения — {df["city_сenters_nearest"].isna().sum()}')
```

Вероятно, что во всех сгруппированных строчках по некоторым населённым пунктам значения в таблице были пропущены, поэтому медиану брать не от чего, соответственно, пропуски остались.
Заполнение оставшихся пропусков общей медианой даст скорее всего не очень адекватные значения.

* первый;
* последний;
* другой.

Затем применим функцию к датафрейму и поместим значения в новый столбец `floor_category`, расположив его рядом со столбцом `floor`.

```python
# функция принимает в качестве аргумента строку дф
# в теле функции этаж квартиры сравнивается с единицей и верхним этажом и возвращает категорию этажа

def floor_category(row):
    
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

df.insert(loc=11, column='floor_category', value=df.apply(floor_category, axis=1))
```

### Удаление редких и выбивающихся значений

Уберём редкие и выбивающиеся значения и опишем обнаруженные особенности.

Здесь также будем руководствоваться правилом о том, что значения, попадающие в размах "усов" диаграммы размаха, и есть статистически значимая выборка (без выбросов).

Чтобы не определять границы "усов" по графикам и не перегружать излишними визуализациями проект, напишем функцию, которая и будет считать границы интервалов нормальных значений.

В переменную `norm_df` скопируем датафрейм, и функцию напишем уже для этой новой таблицы, чтобы последовательно очищать данные от выбивающихся значений в каждой колонке, и таким образом шаг за шагом формировать новый датафрейм со статистически нормальными значениями параметров.

```python
# новый датафрейм
norm_df = df

def stat_sample(row):
    
    # межквартильный размах
    iqr = norm_df[row].quantile(0.75) - norm_df[row].quantile(0.25)
    
    # сравниваем левую границу усов с минимальным значением выборки
    if (norm_df[row].quantile(0.25) - 1.5 * iqr) > norm_df[row].min():
        print(f'Левая граница выборки: {norm_df[row].quantile(0.25) - 1.5 * iqr}')
    else:
        print(f'Левая граница выборки: {norm_df[row].min()}')
    
    # сравниваем правую границу усов с максимальным значением выборки
    if (norm_df[row].quantile(0.75) + 1.5 * iqr) < norm_df[row].max():
        print(f'Правая граница выборки: {norm_df[row].quantile(0.75) + 1.5 * iqr}')
    else:
        print(f'Правая граница выборки: {norm_df[row].max()}')
   

# проверим функцию на столбце days_exposition —
# границы его нормальных значений мы уже считали выше

stat_sample('days_exposition')
```

И теперь начнём обработку таблицы по следующим колонкам: `total_area`, `price_per_meter`, `rooms`, `ceiling_height` и `days_exposition`.

Значения в остальных колонках являются либо категориальными, либо рассчитанными на основе значений в упомянутых колонках, поэтому рассматривать их нет смысла.
Колонку `price_per_meter` предпочитаем оригинальной `last_price`, потому как цена за кв. метр -- более показательный, чем общая площадь, параметр.

### Определение центральной зоны

Выясним, какая область входит в центр. 

Сформируем таблицу `spb_df` объявлений в Санкт-Петербурге.

```python
spb_df = norm_df.query('short_locality == "Санкт-Петербург"').reset_index(drop=True)

display(spb_df.head())
spb_df.columns.get_loc('city_сenters_nearest') # индекс колонки для определения позиции нового столбца
```

В `spb_df` создадим столбец `centre_km` с расстоянием до центра в километрах. После этого посчитаем среднюю цену для каждого километра.

```python
# новый столбец с расстоянием в км
spb_df.insert(loc=20, column='centre_km', value=(spb_df['city_сenters_nearest'] / 1000).round())
```

```python
spb_km_price = spb_df.groupby('centre_km').agg({'price_per_meter': 'mean'}).round()
spb_km_price
```

Построим график зависимости цены от удалённости от центра. По нему определим границу, где график сильно меняется, — это и будет центральная зона.

```python
spb_km_price.plot(title='Зависимость цены кв.метра от удалённости от центра', 
                  kind='bar', grid = True, color='gray', figsize=(12,6));
```

На графике заметен резкий перегиб между 7-ым и 9-ым километром, а дальше цена снижается болле плавно, поэтому можно назвать 8-ой километр границей центральной зоны Санкт-Петербурга.

**Вывод**

Центр Санкт-Петербурга — зона радиусом 8 километров.
