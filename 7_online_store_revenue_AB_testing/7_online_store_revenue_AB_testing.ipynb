- `Hypothesis` — краткое описание гипотезы;

### Приоритизация по ICE

Применим фреймворк _ICE_ для приоритизации гипотез и отсортируем их по убыванию приоритета.

Добавим к таблице `hypothesis` столбец `ice_score`. Результаты ICE для удобства округлим до двух знаков после запятой и отсортируем гипотезы по убыванию параметра.

```python
hypothesis['ice_score'] = (hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']).round(2)
hypothesis[['hypothesis', 'hyp_alias', 'ice_score']].sort_values(by='ice_score', ascending=False)
```

Наиболее перспективные гипотезы по ICE — это:
- скидки в день рождения, 
- новые каналы трафика, 
- добавление формы подписки,
- показ баннера с акциями.

### Приоритизация по RICE

Теперь применим фреймворк _RICE_ для приоритизации гипотез и снова отсортируем их по убыванию приоритета.

```python
hypothesis['rice_score'] = (
    (hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts'])
    .round(2)
)

hypothesis[['hypothesis', 'hyp_alias', 'rice_score']].sort_values(by='rice_score', ascending=False)
```

По RICE наиболее перспективными снова оказались гипотезы:
- добавление формы подписки,
- новые каналы трафика,
- показ баннера с акциями.

Но к ним добавилась — на второе место — гипотеза о добавлении блоков рекомендаций.

## Анализ A/B-теста

### Обзор и подготовка данных

Загрузим данные о визитах и заказах и сохраним их в переменные `visitors` и `orders`.

```python
visitors, orders = (
    pd.read_csv('/datasets/visitors.csv'),
    pd.read_csv('/datasets/orders.csv'),
)
```

- `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста


Дубликатов и пропусков нет, названия столбцов в порядке. Следует изменить типа данных столбца `date`.

Также сразу проверим, корректно ли было деление трафика теста, и посмотрим на диапазон дат.

Тест проводился в течение августа 2019 года, а в группы попало разное число пользователей — нужно вернуться к этому моменту.

```python
display(orders.head(3))
display(f'Количество дубликатов: {orders.duplicated().sum()}')

orders.info()
```

Таблица `orders` содержит следующие колонки:

Дубликатов и пропусков нет. Названия столбцов следует привести к "змеиному регистру", а тип данных столбца `date` изменить на _datetime_.

```python
orders = orders.rename(
    columns={
        'transactionId': 'transaction_id', 
        'visitorId': 'visitor_id'
    }
)
display(orders.columns) # проверка

orders['date'] = orders['date'].map(
    lambda x: dt.datetime.strptime(x, '%Y-%m-%d')
)
```

Выведем и сравним диапазон дат.

```python
display(f'Период проведения теста: с {orders["date"].min()} до {orders["date"].max()}')
```

Даты в обеих таблицах совпадают.

Теперь проверим, нет ли пользователей, которые попали в обе группы, т.е. увидели обе версии страницы.

```python
orders.groupby('visitor_id').agg({'group':'nunique'}).query('group > 1').count()
```

Таких оказалось целых 58! Вполне возможно, что это количество не оказало статистически значимого влияния на результаты теста, но всё-таки очистим данные от "бракованных" пользователей, т.к. если пользователь видит разные версии исследуемой страницы в ходе одного исследования, неизвестно, какая именно повлияла на его решения. Значит, и результаты такого теста нельзя интерпретировать однозначно. 

Сохраним список с их идентификаторами в переменную `broken_visitors` 

```python
broken_visitors = orders.groupby('visitor_id', as_index=False).agg({'group':'nunique'}).query('group > 1')
broken_visitors = broken_visitors['visitor_id']
```

Теперь сформируем таблицу `broken_orders` с заказами, которые совершили пользователи из списка `broken_visitors`.
В ней сгруппируем значения по группе теста и дате заказа, и посчитаем количество транзакций.

```python
broken_orders = (
    orders.query('visitor_id in @broken_visitors')
    .groupby(['group', 'date'], as_index=False)
    .agg({'transaction_id': 'nunique'})
)

broken_orders.rename(columns={'transaction_id':'extra_visit'}, inplace=True)
broken_orders.sample(5)
```

Теперь присоединим таблицу `broken_orders` к таблице `visitors` по колонкам с датой и группой. Пропуски в столбце `extra_visit` заменим нулями.

<div class="alert alert-success">
<b>Комментарий ревьюера ✔️:</b> Всё последовательно </div>

```python
visitors = visitors.merge(broken_orders, how='left', on=['group', 'date']).fillna(0)
visitors.sample(3)
```

И, наконец, вычтем из колонки `visitors` с количеством посещений в каждую дату каждой группой значения из колонки `extra_visit` с посещениями "отбраковынных" пользователей. Сама колонка больше не нужна.

```python
visitors['visitors'] = visitors['visitors'] - visitors['extra_visit'].astype('int')
visitors.drop(['extra_visit'], axis=1, inplace=True)
visitors.sample(3)
```

Снова проверим количество посетителей, попавших в каждую группу теста.

```python
visitors_grouped = visitors.groupby('group').agg({'visitors':'sum'})
visitors_grouped
```

Значения не равны, но для успешного теста количество пользователей в различных группах может различается не более, чем на 1%.

```python
visitors_grouped = visitors_grouped['visitors']

display(f'Количество поситителей в группе B больше количества поситителей ' 
        f'в группе A на {(visitors_grouped[1] / visitors_grouped[0] * 100 - 100).round(2)} %')
```

Также удалим транзакции от "отбракованных" пользователей из таблицы `orders`.

```python
orders = orders[~orders['visitor_id'].isin(broken_visitors)]
```

**Вывод:** загружены данные о посетителях и заказах. В таблицах нет пропусков и явных дубликатов. Типы данных в столбцах с датами изменены на datetime для возможности работы с ними в ходе проекта. Названия столбцов таблиц приведены к "змеиному" регистру.

Данные очищены от пользователей, которые одновременно попали в две группы, и их посещений. Даты теста в обеих таблицах корректны, а разница в количестве пользователей в двух группах не критична для A/B-теста.

###  Кумулятивная выручка по группам

Построим график кумулятивной выручки по группам. Для этого сначала подготовим данные.

```python
# создаем массив уникальных пар значений дат и групп теста
dates_groups = orders[['date','group']].drop_duplicates()

# получаем агрегированные кумулятивные по дням данные о заказах 
orders_agg = (
    dates_groups
    .apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
    .agg(
        {
            'date' : 'max',
            'group' : 'max',
            'transaction_id' : 'nunique',
            'visitor_id' : 'nunique',
            'revenue' : 'sum'
        }
     ), axis=1)
    .sort_values(by=['date','group'])
)

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitors_agg = (
    dates_groups
    .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
    .agg(
        {
            'date' : 'max',
            'group' : 'max',
            'visitors' : 'sum'
        }
     ), axis=1)
    .sort_values(by=['date','group'])
)

# объединяем кумулятивные данные в одной таблице и меняем названия колонок
cumulative_data = orders_agg.merge(visitors_agg, left_on=['date', 'group'], right_on=['date', 'group'])
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
cumulative_data.head(5)
```

Теперь из таблицы `cumulative_data` формируем датафреймы `cumulative_revenue_a` и `cumulative_revenue_b` с кумулятивной выручкой и количеством заказов в каждой из групп А/В-теста.

```python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulative_revenue_a = cumulative_data[cumulative_data['group']=='A'][['date','revenue', 'orders']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulative_revenue_b = cumulative_data[cumulative_data['group']=='B'][['date','revenue', 'orders']]

# график выручки группы А
plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue'], label='A')

# график выручки группы B
plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue'], label='B')

plt.title(f'Кумулятивная выручка по группам', fontsize=18)
plt.legend()

plt.xlabel('дата', fontsize=11)
plt.ylabel('выручка, руб.', fontsize=11);
```

**Вывод**: Выручка в обеих группах почти равномерно увеличивается в течение всего теста. Однако график выручки группы В резко растёт 19 августа. Это может быть внезапным всплеском числа заказов или появлением очень дорогих заказов в выборке.

### Различия в среднем количестве заказов по "сырым" данным

Посчитаем статистическую значимость различий в среднем количестве заказов между группами по «сырым» данным — до удаления аномальных пользователей.

Сформулируем гипотезы: 
- Нулевая гипотеза: различий в среднем количестве заказов между группами нет. 
- Альтернативная: различия в среднем между группами есть.

Создадим переменные `orders_by_users_a` и `orders_by_users_b` со столбцами `visitor_id`, `orders`. В них для пользователей, которые заказывали хотя бы 1 раз, укажем число совершённых заказов.

```python
orders_by_users_a = (
    orders[orders['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': 'nunique'})
)
orders_by_users_a.columns = ['visitor_id', 'orders']

orders_by_users_b = (
    orders[orders['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': 'nunique'})
)
orders_by_users_b.columns = ['visitor_id', 'orders']
```

В случае, когда в данных есть выбросы, — как в нашем — и распределение значений отличается от нормального, вместо алгебраических метрик (которые чувствительны к выбивающимся значениям) применяются непараметрические критерии, в частности *критерий Манна-Уитни* для проверки, что распределения двух выборок не отличаются.

Поэтому подготовим выборки к проверке критерием Манна-Уитни.
 
Объявим переменные `sample_a` и `sample_b`, в которых пользователям из разных групп будет соответствовать количество заказов. Тем, кто ничего не заказал, будут соответствовать нули. К этим выборкам и будем применять критерий Манна-Уитни.

```python
sample_a = pd.concat(    # объединение последовательностей
    [
        orders_by_users_a['orders'],   # список с количеством совершённых заказов каждым пользователем
        pd.Series(                     # + список
            0,                         # с нулями
            index=np.arange(           # длина списка = количество посетителей, которые ничего не заказали
                                       # т.е. все посетители в группе минус заказавшие
                visitors.query('group=="A"')['visitors'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sample_b = pd.concat(
    [
        orders_by_users_b['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors.query('group=="B"')['visitors'].sum() - len(orders_by_users_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

Теперь применим критерий и получим *p-value*.
А также посчитаем относительный прирост среднего числа заказов группы B.

```python
display('Значение p-value = {0:.3f}'.format(stats.mannwhitneyu(sample_a, sample_b)[1]))

# среднее из выборки sample_ есть среднее число заказов на пользователя в группе
display('Относительный выигрыш группы В равен {0:.3f}'.format(sample_b.mean() / sample_a.mean() - 1))
```

**Вывод**: p-value оказлось меньше 0.05. Значит, нулевую гипотезу о том, что статистически значимых различий в среднем числе заказов между группами нет, следует отвергнуть и принять альтернативную: *различия в средних между группами есть*.

Относительный прирост среднего количества заказов в группе B равен целым 16 %.

### Различия в среднем чеке заказа по "сырым" данным

Теперь посчитаем статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным.

- Нулевая гипотеза: различий в среднем чеке между группами нет;
- Альтернативная гипотеза: различия в среднем чеке между группами есть.

Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию mannwhitneyu() данные о выручке с заказов. И найдём относительные различия в среднем чеке между сегментами.

```python
display('Значение p-value = {0:.3f}'
        .format(
            stats.mannwhitneyu(
                orders[orders['group']=='A']['revenue'],
                orders[orders['group']=='B']['revenue'])
            [1]
        )
       )

display('Относительный выигрыш группы В равен {0:.3f}'
        .format(
            orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1
        )
       )
```

**Вывод**: p-value больше 0.05. Значит, причин отвергать нулевую гипотезу и считать, что в среднем чеке есть различия, нет. Видимо, показатель относительного различия в чеках групп — почти 29 % — получен случайным образом.

### Различия в среднем количестве заказов по "очищенным" данным

Посчитаем статистическую значимость различий в среднем количестве заказов между группами по «очищенным» данным.

Для этого сперва "очистим" данные от аномальных польщователей, которые совершали много или дорогие заказы.

Примем за аномальных пользователей тех, кто совершил больше 2 заказов или совершил заказ дороже 30 000 рублей. Так мы уберём 1 % пользователей с наибольшим числом заказов и от 1% до 5% пользователей с дорогими заказами. Сделаем срезы пользователей с числом заказов больше 2 — `users_with_many_orders` — и пользователей, совершивших заказы дороже 30 000 — `users_with_expensive_orders`. И затем объединим их в таблице `abnormal_users`.

```python
too_many_orders = 2
too_expensive_order = 30000

users_with_many_orders = orders_by_users[orders_by_users['orders'] > too_many_orders]['visitor_id']

users_with_expensive_orders = orders[orders['revenue'] > too_expensive_order]['visitor_id']

# склеиваем списки пользователей
abnormal_users = (
    pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
    .drop_duplicates() # на случай, если пользователь угодил в оба списка
    .sort_values()
)
display(abnormal_users.head(5))
display(f'В списке аномальных пользователей {abnormal_users.shape[0]} человек')
```

Гипотезы те же:

- Нулевая гипотеза: различий в среднем количестве заказов между группами нет.
- Альтернативная: различия в среднем между группами есть.

Снова подготовим выборки для проверки критерием Манна-Уитни, исключив из них заказы аномальных пользователей.

```python
sample_a_filtered = pd.concat(
    [
        orders_by_users_a[
            np.logical_not(orders_by_users_a['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors.query('group=="A"')['visitors'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sample_b_filtered = pd.concat(
    [
        orders_by_users_b[
            np.logical_not(orders_by_users_b['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitors.query('group=="B"')['visitors'].sum() - len(orders_by_users_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

И узнаем, как действия исключённых пользователей повлияли на результаты теста. Посчитаем статистическую значимость различий в среднем количестве заказов между группами теста по очищенным данным — применим статистический критерий Манна-Уитни к полученным выборкам:

```python
display('Значение p-value = {0:.3f}'.format(stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]))

display('Относительный выигрыш группы В равен {0:.3f}'
        .format(sample_b_filtered.mean() / sample_a_filtered.mean() - 1))
```

**Вывод**: Результаты по среднему количеству заказов практически не изменились, а значит *различия в средних между группами есть*.

Относительный прирост среднего количества заказов в группе B увеличился до 18,7 %.

### Различия в среднем чеке заказа по "очищенным" данным

Проверим, изменились ли результаты по среднему чеку.

Аналогично, гипотезы:

- Нулевая гипотеза: различий в среднем чеке между группами нет;
- Альтернативная гипотеза: различия в среднем чеке между группами есть.

Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным.


```python
display('Значение p-value = {0:.3f}'
        .format(
            stats.mannwhitneyu(
                orders[
                    np.logical_and(
                        orders['group']=='A',
                        np.logical_not(orders['visitor_id'].isin(abnormal_users))
                    )
                 ]['revenue'],
                orders[
                    np.logical_and(
                        orders['group']=='B',
                        np.logical_not(orders['visitor_id'].isin(abnormal_users))
                    )
                 ]['revenue']
            )
            [1]
        )
       )

display('Относительный выигрыш группы В равен {0:.3f}'
        .format(
            orders[
                np.logical_and(
                    orders['group']=='B',
                    np.logical_not(orders['visitor_id'].isin(abnormal_users))
                )
            ]['revenue'].mean()
            / orders[
                np.logical_and(
                    orders['group']=='A',
                    np.logical_not(orders['visitor_id'].isin(abnormal_users))
                )
            ]['revenue'].mean()
            - 1
        )
       )
```

**Вывод**: p-value всё ещё больше 0,05 — всё ещё не отвергаем нулевую гипотезу: различий в среднем чеке между группами нет.

Но теперь и оценённое относительное различие между средними чеками групп упало с 29 % до 2,7 % — это подтверждает, что в сырых данных оно было случайным и вызванным аномально дорогими заказами в выборке.

### Результат теста

- **Нет статистически значимого различия по среднему чеку** между группами ни по «сырым», ни по данным после фильтрации аномалий. На значительное относительное различие между чеками групп на "сырых" данных повлияли заказы с аномально высокой стоимостью;
- **Есть статистически значимое различие по среднему количеству заказов** между группами и в «сырых» данных, и в "очищенных". После фильтрации аномалий относительный прирост среднего количества заказов в группе B даже увеличился на 2,7 процентных пункта и составил 18,7 %;
- На графиках **различия среднего чека**, результаты группы B начинают резко превосходить результаты группы А только после резкого скачка, вызванного аномально дорогим заказом пользователя, попавшего в группу В, после чего их тренды стремятся **стабилизироваться вокруг близких значений**;
- На графики **различия конверсии между группами** выбросы сильно не повлияли: кривые колеблются в начале теста, но довольно быстро выравниваются на отметках ~2,5 % (группа А) и ~2,9 % (группа В) (для "сырых" данных). График относительного изменения кумулятивной конверсии ещё испытывает некоторые флуктуации даже в конце теста, но уже понятно, что значения стабилизируются в ближайшие дни, и **результаты группы В лучше, чем у группы А**.

Исходя из этого, **следует остановить тест и признать его успешным**: **конверсию в покупки удалось увеличить на 18 %**, **к увеличению среднего чека изменение не привело**, но и хуже не стало.
