Следующая колонка с пропусками — `year_of_release`.

```python
data[data['year_of_release'].isna()].head()
```

Лёгкий гуглинг даёт основание полагать, что год, указывающийся в названиях некоторых игр — как правило, следующий за годом выпуска. Можно попробовать заполнить пропуски в строках с играми, в названиях которых указан год.
Заменять год выхода игры на одной платформе годом её выхода на другой не будем, т.к. даты таких релизов могут отличаться на несколько лет.

Напишем функцию, которая "вытащит" год из названия.

```python
def year_out_of_name(row):
    
    game_name = row['name']
    game_year = row['year_of_release']
    
    if pd.isna(game_year):                      # если в столбце с годом выпуска пропуск;
        for word in game_name.split(sep=' '):   # делим строку из названия на отдельные слова по пробелу;
            if word.isdigit():                  # проверяем, является ли каждое "слово" числом;
                word = int(word)                # преобразуем str в int;
                if 1980 <= word <= 2016:        # проверяем, входит ли оно в диапазон лет наших данных;
                    return word - 1             # возвращаем предыдущий год;
    return game_year                            # если пропуска года не было, возвращаем обратно год.
```

```python
display(f'количество пропусков до применения функции: {data["year_of_release"].isna().sum()}')

# применяем функцию к столбцу с годом выпуска:
data['year_of_release'] = data.apply(year_out_of_name, axis=1) 

display(f'количество пропусков после применения функции: {data["year_of_release"].isna().sum()}')
```

Заполнили целых 15 пропусков! 😅 (probably not worth it)

Значения для остальных взять особо негде, поэтому их просто заполним, например, 2049-ым годом (и будем отфильтровывать датафреймы по последнему, 2016-ому, году при необходимости), чтобы сделать возсожным изменение типа данных в столбце `year_of_release` на _int_ (тут не вижу смысла приводить к типу данных _datetime_, потому что в нашем распоряжении только год).

```python
data['year_of_release'] = data['year_of_release'].fillna(2049).astype('int32')
```

Акроним служит для отметки неясностей или пропусков, которые надо заполнить, в информации требований.

То есть по сути строки с этим значение можно приравнять к пропускам, заполнить те и другие одинаково. В то же время среди уникальных значений в столбце с оценкой пользователей встречается оценка _0_, поэтому нулями заполнять нельзя — смешаются реальные и "индикаторные" значения. 

Заполним пропуски и заменим значения _tbd_ значением _-1_: таким образом, сделаем возможным преобразование к числовому типу данных, визуально пометим строки с неизвестными значениями, обеспечим лёгкость и правильность арифметических операций с оценками пользователей в дальнейшем — просто будем обрабатывать лишь положительные значения в столбце, а также пики на графиках распределения в значении _-1_ не будут смешиваться с реальными значениями.

```python
data['user_score'] = data['user_score'].fillna(-1).replace('tbd', -1).astype('float32')
```

Следующий столбец с пропусками — `critic_score`. Хоть минимальное числовое значение в этом столбце — 13, и нулями данные мы не исказим, но пропуски по аналогии с пользовательскими оценками заменим здесь значением _-1_.


```python
data['critic_score'] = data['critic_score'].fillna(-1)
```

## Исследовательный анализ данных

### Выпуск игр в разные годы

Посмотрим, сколько игр выпускалось в разные годы. Построим гистограмму, не будем учитывать на ней 2049 год, которым заполнены пропуски.

```python
data[data['year_of_release'] != 2049]['year_of_release'].plot(
    kind='hist', 
    title='Количество выпущенных игр в год',
    bins=37, 
    grid=True,
    color='black',
    alpha=0.5);
```

Видим по гистограмме, что начало развития индустрии пришлось на 80-е годы (это самые ранние наши данные), затем произошёл стремительный рост с середины 90-х, пик в 2008-ом, а дальше спад. Ввидимо, популярность набирали мобильные игры. Или же причина нового нисходящего тренда в том, что производители игр пошли по пути перехода от количественного показателя к качественному.

Заметен странный пик на 2016-ом году. Рассмотрим ещё изменение суммарных продаж по годам.

```python
data[data['year_of_release'] != 2049]\
    .groupby('year_of_release')\
    .agg({'all_sales':'sum'}).plot(
        grid=True,
        color='darkslategrey',
        linewidth=4
    )
plt.title('Суммарные продажи на всех платформах по годам', fontsize=18)
plt.xticks(range(1980,2017), rotation=45);
```

В целом, график продаж соотносится с графиком количества выпущенных игр, но никакого резкого роста в 2016-ом нет. Пользователь стал так избирателен, что даже наращивание ассортимента не спасает от падения продаж?

Для прогнозирования событий в будущем важно выделить для анализа временной период, который сбалансированно отразит актуальные и исторические данные. Актуальные данные учитывают текущие изменения на рынке. Но исторические данные, в свою очередь, тоже могут быть крайне полезны для прогнозирования, возможно, показывать некую модель поведения, но главное — затрагивать исторические периоды, характеризующиеся разными трендами (чтобы не получилось как с моделью VaR 😅).

### Изменение продаж по платформам

Посмотрим, как вообще менялись продажи по платформам.
Построим точечную диаграмму, которая отразит суммарные продажи игр на каждой из платформ с распределением по годам.

```python
sns.scatterplot(x='year_of_release', y='platform',
                hue='platform',
                size='all_sales',
                data=data,
                sizes=(30, 700))
plt.xlim(1980, 2017)
plt.title('Изменение количества продаж по платформам', fontsize=18)
plt.xticks(range(1980,2017), rotation=45)
plt.grid(True, alpha=.3)
plt.legend(bbox_to_anchor=(-0.1, 1));
```

График получился мелкомасштабным, но уже по нему видно, что у платформ, как правило, есть некоторый "срок жизни", оканчивающийся спадом продаж. Пик продаж на некотрых платформах приходится прямо на их старт, на других же случается в середине их "жизненного цикла". Наименее популярные платформы существовали лишь в пределах одного года!

Теперь выберем платформы с наибольшими суммарными продажами. Для этого сформируем сводную таблицу, в которой отсортируем платформы по убыванию суммарных продаж.

```python
top_platforms = data[data['year_of_release'] != 2049]\
    .groupby(['platform'])\
    .agg({'all_sales':'sum'})\
    .sort_values(by='all_sales', ascending=False)

top_platforms
```

Отсечём перечень по платформе *XOne* — со следующей платформой в списке большой разрыв в суммарных продажах (почти в два раза) — и сформируем список из отобранных платформ.

```python
top_platforms = list(top_platforms.head(18).index)
top_platforms
```

Построим аналогичное распределение продаж по годам уже лишь с выбранными платформами.

```python
data_top_platform = data.query('platform == @top_platforms')

sns.scatterplot(x='year_of_release', y='platform',
                hue='platform',
                size='all_sales',
                data=data_top_platform,
                sizes=(30, 700))
plt.xlim(1980, 2017)
plt.title('Изменение количества продаж по платформам с наибольшими суммарными продажами', fontsize=18)
plt.xticks(range(1980,2017), rotation=45)
plt.grid(True, alpha=.3)
plt.legend(bbox_to_anchor=(-0.1, 1));
```

На таком усечённом графике уже проще оценить, какие платформы когда существовали и как менялись суммарные продажи игр на этих платформах.

Чтобы понять, за какой характерный срок появляются новые и исчезают старые платформы, рассчитаем "время жизни" каждой из них.

### Влияние отзывов  на продажи

Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков по диаграмме рассеяния.

Напишем функцию, в теле которой будет строиться два графика: зависимость количества проданных копий игр от оценок пользователей и оценок критиков на каждой отдельно взятой платформе. Тут же рассчитаем и нанесём на графики коэффициенты корреляции этих величин.

```python
def score_impact(platf):
    
    # две таблицы, чтобы сохранить максимум данных: в строке может присутствовать только одна из оценок:
    # избавляемся от индикаторных значений пропусков рейтингах (-1)
    user_score_data = actual_data.query('user_score >= 0 and platform == @platf').reset_index(drop=True)
    critic_score_data = actual_data.query('critic_score >= 0  and platform == @platf').reset_index(drop=True)
    
    fig, axis = plt.subplots(1,2)
    
    # параметры текстовой плашки
    props = dict(boxstyle='round', facecolor='thistle', alpha=0.5)
    
    # диаграмма рассеяния для оценок пользователей
    sns.scatterplot(
        data=user_score_data,
        x='user_score',
        y='all_sales',
        hue = 'user_score',
        size = 'user_score',
        sizes = (10,100), 
        ax=axis[0]
    ).set(title='оценка пользователей — продажи')
    
    # коэффициент корреляции 'оценка пользователей — продажи'
    text_1 = f'Коэф.корреляции: {user_score_data["all_sales"].corr(user_score_data["user_score"]):.2f}'
    plt.text(0.02, 0.5, text_1, 
             horizontalalignment='center', verticalalignment='center', 
             transform=ax.transAxes, bbox=props, fontsize='large')

    # диаграмма рассеяния для оценок критиков
    sns.scatterplot(
        data=critic_score_data,
        x='critic_score',
        y='all_sales',
        hue = 'critic_score',
        size = 'critic_score',
        sizes = (10,100), 
        ax=axis[1]
    ).set(title='оценка критиков — продажи')
    
    # коэффициент корреляции 'оценка критиков — продажи'
    text_2 = f'Коэф.корреляции: {critic_score_data["all_sales"].corr(critic_score_data["critic_score"]):.2f}'
    plt.text(0.55, 0.5, text_2, 
             horizontalalignment='center', verticalalignment='baseline', 
             transform=ax.transAxes, bbox=props, fontsize='large')
    
    # одинаковые границы графиков для оси х
    axis[0].set_xlim(0,10)
    axis[1].set_xlim(0,100)
    
    fig.suptitle(f'Платформа {platf}', fontsize=18)
```

И применим написанную функцию в цикле к каждой из шести отобранных популярных платформ:

```python
for i in top_6_platforms:
    
    score_impact(i)
```

По построенным зависимостям можно сделать вывод о том, что ни в одном из шести случаев _оценки пользователей не коррелируют с продажами игр_. Пользователи склонны оценивать игры в 5—8 баллов, действительно часто высоко оценивают успешные релизы, однако чёткой зависимости тут не вырисовывается.

Сравнительно более заметная, но всё же довольно _слабая положительная корреляция существует между продажами оценками критиков_. Наиболее выражена она на платформах _XOne_ и _PS4_, наименее (практически отсутствует) — на платформе _PC_.

Также стоит отметить относительную солидарность между пользователями и критиками на платформе _3DS_, распространённой преимущественно в японском регионе.

Таким образом, можно сказать, что оценки критиков можно (но осторожно) использовать для прогнозирования успеха выпускаемых игр, в то время как оценки пользователей для этих целей практически бесполезны.

## Составление портрета пользователя каждого региона

Определим некоторые характерные особенности для пользователя каждого региона (NA, EU, JP).

### Самые популярные платформы

Выявим топ-5 популярных платформ в каждом из регионов. Сравнивать будем доли продаж на каждой платформе в каждом регионе.

```python
def top_region_share(index, reg, name_reg, bar_color): 
    
    # функция принимает в качестве аргументов:
            # индекс для группировки, 
            # регион продаж, 
            # название региона для подписи графика и 
            # цвет для столбцов диаграммы;
            
    # формирует сводную таблицу по индексу и региону за актуальный период:
       
    region_data = data.query('2013 <= year_of_release <= 2016')\
    .pivot_table(index=index, values=reg, aggfunc='sum')
    
    # добавляет в таблицу столбец с долей продаж для выбранного индекса в регионе 
    # от суммарных продаж в регионе в процентах:
    
    region_data[reg+'_share'] = region_data[reg] / region_data[reg].sum() * 100
    
    # сортирует таблицу и строит столбчатую диаграмму топ-5 по переданному в функцию индексу в регионе:
    
    region_data\
    .sort_values(by=reg)\
    .tail()\
    .plot(
        kind='barh', 
        y=reg+'_share',
        color=bar_color,
        legend=True,
        figsize=(14,4)
    )
    plt.title(f'Топ-5 в {name_reg}', fontsize=18)
    plt.xlim(0,50) # одинаковый масштаб для графиков: от 0 до 50 %
    plt.xlabel('sales_share in %')
```

```python
regions = {'na_sales': 'NA', 'eu_sales': 'EU', 'jp_sales': 'JP'} # словарь с регионами

# и применяем функцию к каждому региону по очереди. Регионы перебираем из словаря:

for key, value in regions.items():
    top_region_share('platform', key, value, 'burlywood')
```

В Северной Америке (NA) и Европе (EU) распределение схожее: платформа _PS4_ превосходит по популярности остальные с заметным отрывом: в NA на её долю приходится около 25 % продаж, а в EU все 36 %. Топ в этих регионах составляют одни и те платформы — _PS3, XOne, X360, 3DS_ , — лишь незначительно варьируясь по местам. Среди европейских пользователей платформа _PS3_ гораздо популярнее, чем у североамериканских, а платформа _3DS_ и там и там список замыкает (около 8 % продаж).

Совсем другая ситуация в Японии (JP): тут с драматическим отрывом вперёд вырывается платформа _3DS_ (почти 50 %!). Остальные входящие в список наиболее популярных — _PS3, PSV, PS4 и WiiU_ — отстают друг от друга незначительно — продпжи от 8 до 17 %. ## Проверка гипотез

### Различие пользовательских рейтингов на разных платформах

Проверим, различаются ли средние пользовательские рейтинги платформ _Xbox One_ и _PC_.

Для этого сначала сформулируем *нулевую гипотезу* **Н0** и *альтернативную гипотезу* **Н1**:

**Н0**: средний пользовательский рейтинг платформы _Xbox One_ ***равен*** среднему пользовательскому рейтингу платформы _PC_.

**Н1**: средний пользовательский рейтинг платформы _Xbox One_ ***отличается*** от среднего пользовательского рейтинга платформы _PC_.

Сравниваемые выборки *независимы* между собой, поэтому будем использовать *t-критерий Стьюдента для несвязанных выборок*. С его помощью можно оценить, насколько различаются их средние.
Т.к. формула t-критерия включает в себя средние значения выборок, то критерий может давать неадекватные результаты при наличии выбросов. В разделе Предобработка данных мы уже убедились, что в колонке нет значений-выбросов, однако перед сравнением выборок нужно будет отсеять индикаторные значения пропусков (= -1) в столбце `user_score`.

Сформируем массивы `data_xone` и `data_pc` со значениями пользовательских рейтингов для использования их в функции далее.

```python
data_xone = data.query('2013 <= year_of_release <= 2016 and platform == "XOne" and user_score >= 0')['user_score']
data_pc = data.query('2013 <= year_of_release <= 2016 and platform == "PC" and user_score >= 0')['user_score']
```

Чтобы определить значение параметра `equal_var` функции `st.ttest_ind`, сравним на всякий случай размеры выборок. Дисперсии в них значительно отличаться не могут, так как значения рейтинга ограничены диапазоном от 0 до 10 баллов.

```python
display(len(data_xone), len(data_pc))
```

Размеры выборок отличаются, но незначительно — будем использовать значение параметра `equal_var` по умолчанию.

И можно наконец проверять *гипотезу о равенстве средних* двух генеральных совокупностей по взятым из них выборкам. Критерий значимости `alpha` зададим стандартный — 5 %.

```python
alpha = 0.05

results_1 = st.ttest_ind(data_xone, data_pc)
display(f'значение p-value: {results_1.pvalue}')
results_1.pvalue < alpha
```

Полученное значение p-value _больше_ критерия значимости alpha и говорит о том, что вероятность получить такое или большее различие случайно — около 14 %, следовательно, между выборками отсутствует значимое различие, а значит нулевая гипотеза не может быть отвергнута, т.е. средние пользовательские рейтинги платформ _Xbox One_ и _PC_ **одинаковые**.

### Различие пользовательских рейтингов в разных жанрах

Средние пользовательские рейтинги жанров Action и Sports разные.

Проверим, отличаются ли средние пользовательские рейтинги жанров _Action_ и _Sports_.

Сформулируем гипотезы:

**Н0**: средний пользовательский рейтинг у игр жанра _Action_ ***равен*** среднему пользовательскому рейтингу игр жанра _Sports_.

**Н1**: средний пользовательский рейтинг у игр жанра _Action_ ***отличается*** от среднего пользовательского рейтинга игр жанра _Sports_.

Эти выборки также *независимы*, поэтому используем всё тот же *t-критерий Стьюдента для несвязанных выборок* и проверяем *гипотезу о равенстве средних* двух генеральных совокупностей по взятым из них выборкам.

Формируем массивы `data_action` и `data_sports`:

```python
data_action = data.query('2013 <= year_of_release <= 2016 and genre == "Action" and user_score >= 0')['user_score']
data_sports = data.query('2013 <= year_of_release <= 2016 and genre == "Sports" and user_score >= 0')['user_score']
```

И проверим размер выборок:

```python
display(len(data_action), len(data_sports))
```

На этот раз размеры выборок отличаются более, чем в два раза, поэтому применим параметр `equal_var` со значением `False`.

```python
results_1 = st.ttest_ind(data_action, data_sports, equal_var=False)
display(f'значение p-value: {results_1.pvalue}')
results_1.pvalue < alpha
```

Полученное значение p-value значительно _меньше_ критерия значимости `alpha`, следовательно, между выборками существует значимое различие, поэтому отвергаем нулевую гипотезу и принимаем альтернативную: cредние пользовательские рейтинги жанров _Action_ и _Sports_ **разные**.

Посмотрим также на разницу в средних.

```python
data_action.mean() / data_sports.mean() - 1
```

Средние отличаются на 30 %, что, действительно, довольно много.
