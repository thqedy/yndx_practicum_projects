–õ—ë–≥–∫–∏–π –≥—É–≥–ª–∏–Ω–≥ –¥–∞—ë—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–µ –ø–æ–ª–∞–≥–∞—Ç—å, —á—Ç–æ –≥–æ–¥, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π—Å—è –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –∏–≥—Ä ‚Äî –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, —Å–ª–µ–¥—É—é—â–∏–π –∑–∞ –≥–æ–¥–æ–º –≤—ã–ø—É—Å–∫–∞. –ú–æ–∂–Ω–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –∑–∞–ø–æ–ª–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç—Ä–æ–∫–∞—Ö —Å –∏–≥—Ä–∞–º–∏, –≤ –Ω–∞–∑–≤–∞–Ω–∏—è—Ö –∫–æ—Ç–æ—Ä—ã—Ö —É–∫–∞–∑–∞–Ω –≥–æ–¥.
–ó–∞–º–µ–Ω—è—Ç—å –≥–æ–¥ –≤—ã—Ö–æ–¥–∞ –∏–≥—Ä—ã –Ω–∞ –æ–¥–Ω–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –≥–æ–¥–æ–º –µ—ë –≤—ã—Ö–æ–¥–∞ –Ω–∞ –¥—Ä—É–≥–æ–π –Ω–µ –±—É–¥–µ–º, —Ç.–∫. –¥–∞—Ç—ã —Ç–∞–∫–∏—Ö —Ä–µ–ª–∏–∑–æ–≤ –º–æ–≥—É—Ç –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ª–µ—Ç.

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è "–≤—ã—Ç–∞—â–∏—Ç" –≥–æ–¥ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è.

```python
def year_out_of_name(row):
    
    game_name = row['name']
    game_year = row['year_of_release']
    
    if pd.isna(game_year):                      # –µ—Å–ª–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ —Å –≥–æ–¥–æ–º –≤—ã–ø—É—Å–∫–∞ –ø—Ä–æ–ø—É—Å–∫;
        for word in game_name.split(sep=' '):   # –¥–µ–ª–∏–º —Å—Ç—Ä–æ–∫—É –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞ –ø–æ –ø—Ä–æ–±–µ–ª—É;
            if word.isdigit():                  # –ø—Ä–æ–≤–µ—Ä—è–µ–º, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –∫–∞–∂–¥–æ–µ "—Å–ª–æ–≤–æ" —á–∏—Å–ª–æ–º;
                word = int(word)                # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º str –≤ int;
                if 1980 <= word <= 2016:        # –ø—Ä–æ–≤–µ—Ä—è–µ–º, –≤—Ö–æ–¥–∏—Ç –ª–∏ –æ–Ω–æ –≤ –¥–∏–∞–ø–∞–∑–æ–Ω –ª–µ—Ç –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö;
                    return word - 1             # –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –ø—Ä–µ–¥—ã–¥—É—â–∏–π –≥–æ–¥;
    return game_year                            # –µ—Å–ª–∏ –ø—Ä–æ–ø—É—Å–∫–∞ –≥–æ–¥–∞ –Ω–µ –±—ã–ª–æ, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–±—Ä–∞—Ç–Ω–æ –≥–æ–¥.
```

```python
display(f'–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –¥–æ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏: {data["year_of_release"].isna().sum()}')

# –ø—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ —Å—Ç–æ–ª–±—Ü—É —Å –≥–æ–¥–æ–º –≤—ã–ø—É—Å–∫–∞:
data['year_of_release'] = data.apply(year_out_of_name, axis=1) 

display(f'–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏: {data["year_of_release"].isna().sum()}')
```

–ó–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –≤–∑—è—Ç—å –æ—Å–æ–±–æ –Ω–µ–≥–¥–µ, –ø–æ—ç—Ç–æ–º—É –∏—Ö –ø—Ä–æ—Å—Ç–æ –∑–∞–ø–æ–ª–Ω–∏–º, –Ω–∞–ø—Ä–∏–º–µ—Ä, 2049-—ã–º –≥–æ–¥–æ–º (–∏ –±—É–¥–µ–º –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤—ã–≤–∞—Ç—å –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –ø–æ –ø–æ—Å–ª–µ–¥–Ω–µ–º—É, 2016-–æ–º—É, –≥–æ–¥—É –ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏), —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –≤–æ–∑—Å–æ–∂–Ω—ã–º –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–µ `year_of_release` –Ω–∞ _int_ (—Ç—É—Ç –Ω–µ –≤–∏–∂—É —Å–º—ã—Å–ª–∞ –ø—Ä–∏–≤–æ–¥–∏—Ç—å –∫ —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö _datetime_, –ø–æ—Ç–æ–º—É —á—Ç–æ –≤ –Ω–∞—à–µ–º —Ä–∞—Å–ø–æ—Ä—è–∂–µ–Ω–∏–∏ —Ç–æ–ª—å–∫–æ –≥–æ–¥).

```python
data['year_of_release'] = data['year_of_release'].fillna(2049).astype('int32')
```

> TBD - –∞–±–±—Ä–µ–≤–∏–∞—Ç—É—Ä–∞ –æ—Ç –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ To Be Determined (–±—É–¥–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–æ) –∏–ª–∏ To Be Decided (–±—É–¥–µ—Ç —Ä–µ—à–µ–Ω–æ). –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –µ—Å–ª–∏ –∫–∞–∫–∞—è-—Ç–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –µ—â–µ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –∏–ª–∏ —Ä–µ—à–µ–Ω–∏–µ –ø–æ –≤–æ–ø—Ä–æ—Å—É –Ω–µ –ø—Ä–∏–Ω—è—Ç–æ. –ê–∫—Ä–æ–Ω–∏–º —Å–ª—É–∂–∏—Ç –¥–ª—è –æ—Ç–º–µ—Ç–∫–∏ –Ω–µ—è—Å–Ω–æ—Å—Ç–µ–π –∏–ª–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–¥–æ –∑–∞–ø–æ–ª–Ω–∏—Ç—å, –≤ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–π.

–¢–æ –µ—Å—Ç—å –ø–æ —Å—É—Ç–∏ —Å—Ç—Ä–æ–∫–∏ —Å —ç—Ç–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ –º–æ–∂–Ω–æ –ø—Ä–∏—Ä–∞–≤–Ω—è—Ç—å –∫ –ø—Ä–æ–ø—É—Å–∫–∞–º, –∑–∞–ø–æ–ª–Ω–∏—Ç—å —Ç–µ –∏ –¥—Ä—É–≥–∏–µ –æ–¥–∏–Ω–∞–∫–æ–≤–æ. –í —Ç–æ –∂–µ –≤—Ä–µ–º—è —Å—Ä–µ–¥–∏ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ —Å –æ—Ü–µ–Ω–∫–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –æ—Ü–µ–Ω–∫–∞ _0_, –ø–æ—ç—Ç–æ–º—É –Ω—É–ª—è–º–∏ –∑–∞–ø–æ–ª–Ω—è—Ç—å –Ω–µ–ª—å–∑—è ‚Äî —Å–º–µ—à–∞—é—Ç—Å—è —Ä–µ–∞–ª—å–Ω—ã–µ –∏ "–∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω—ã–µ" –∑–Ω–∞—á–µ–Ω–∏—è. 

–ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—Å–∫–∏ –∏ –∑–∞–º–µ–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è _tbd_ –∑–Ω–∞—á–µ–Ω–∏–µ–º _-1_: —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å–¥–µ–ª–∞–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫ —á–∏—Å–ª–æ–≤–æ–º—É —Ç–∏–ø—É –¥–∞–Ω–Ω—ã—Ö, –≤–∏–∑—É–∞–ª—å–Ω–æ –ø–æ–º–µ—Ç–∏–º —Å—Ç—Ä–æ–∫–∏ —Å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –æ–±–µ—Å–ø–µ—á–∏–º –ª—ë–≥–∫–æ—Å—Ç—å –∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π —Å –æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º ‚Äî –ø—Ä–æ—Å—Ç–æ –±—É–¥–µ–º –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –ª–∏—à—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ, –∞ —Ç–∞–∫–∂–µ –ø–∏–∫–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ –∑–Ω–∞—á–µ–Ω–∏–∏ _-1_ –Ω–µ –±—É–¥—É—Ç —Å–º–µ—à–∏–≤–∞—Ç—å—Å—è —Å —Ä–µ–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏.

–°—Ñ–æ—Ä–º–∏—Ä—É–µ–º –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü `all_sales` —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –≤–æ –≤—Å–µ—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö, –¥–æ–±–∞–≤–∏–º –µ–≥–æ –≤ —Ç–∞–±–ª–∏—Ü—É.

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –í—ã–ø—É—Å–∫ –∏–≥—Ä –≤ —Ä–∞–∑–Ω—ã–µ –≥–æ–¥—ã

–ü–æ—Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ –∏–≥—Ä –≤—ã–ø—É—Å–∫–∞–ª–æ—Å—å –≤ —Ä–∞–∑–Ω—ã–µ –≥–æ–¥—ã. –ü–æ—Å—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É, –Ω–µ –±—É–¥–µ–º —É—á–∏—Ç—ã–≤–∞—Ç—å –Ω–∞ –Ω–µ–π 2049 –≥–æ–¥, –∫–æ—Ç–æ—Ä—ã–º –∑–∞–ø–æ–ª–Ω–µ–Ω—ã –ø—Ä–æ–ø—É—Å–∫–∏.

```python
data[data['year_of_release'] != 2049]['year_of_release'].plot(
    kind='hist', 
    title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –∏–≥—Ä –≤ –≥–æ–¥',
    bins=37, 
    grid=True,
    color='black',
    alpha=0.5);
```

–í–∏–¥–∏–º –ø–æ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ, —á—Ç–æ –Ω–∞—á–∞–ª–æ —Ä–∞–∑–≤–∏—Ç–∏—è –∏–Ω–¥—É—Å—Ç—Ä–∏–∏ –ø—Ä–∏—à–ª–æ—Å—å –Ω–∞ 80-–µ –≥–æ–¥—ã (—ç—Ç–æ —Å–∞–º—ã–µ —Ä–∞–Ω–Ω–∏–µ –Ω–∞—à–∏ –¥–∞–Ω–Ω—ã–µ), –∑–∞—Ç–µ–º –ø—Ä–æ–∏–∑–æ—à—ë–ª —Å—Ç—Ä–µ–º–∏—Ç–µ–ª—å–Ω—ã–π —Ä–æ—Å—Ç —Å —Å–µ—Ä–µ–¥–∏–Ω—ã 90-—Ö, –ø–∏–∫ –≤ 2008-–æ–º, –∞ –¥–∞–ª—å—à–µ —Å–ø–∞–¥. –í–≤–∏–¥–∏–º–æ, –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –Ω–∞–±–∏—Ä–∞–ª–∏ –º–æ–±–∏–ª—å–Ω—ã–µ –∏–≥—Ä—ã. –ò–ª–∏ –∂–µ –ø—Ä–∏—á–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –Ω–∏—Å—Ö–æ–¥—è—â–µ–≥–æ —Ç—Ä–µ–Ω–¥–∞ –≤ —Ç–æ–º, —á—Ç–æ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª–∏ –∏–≥—Ä –ø–æ—à–ª–∏ –ø–æ –ø—É—Ç–∏ –ø–µ—Ä–µ—Ö–æ–¥–∞ –æ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –∫ –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–º—É.

–í —Ü–µ–ª–æ–º, –≥—Ä–∞—Ñ–∏–∫ –ø—Ä–æ–¥–∞–∂ —Å–æ–æ—Ç–Ω–æ—Å–∏—Ç—Å—è —Å –≥—Ä–∞—Ñ–∏–∫–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –∏–≥—Ä, –Ω–æ –Ω–∏–∫–∞–∫–æ–≥–æ —Ä–µ–∑–∫–æ–≥–æ —Ä–æ—Å—Ç–∞ –≤ 2016-–æ–º –Ω–µ—Ç. –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å—Ç–∞–ª —Ç–∞–∫ –∏–∑–±–∏—Ä–∞—Ç–µ–ª–µ–Ω, —á—Ç–æ –¥–∞–∂–µ –Ω–∞—Ä–∞—â–∏–≤–∞–Ω–∏–µ –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞ –Ω–µ —Å–ø–∞—Å–∞–µ—Ç –æ—Ç –ø–∞–¥–µ–Ω–∏—è –ø—Ä–æ–¥–∞–∂?

–î–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è —Å–æ–±—ã—Ç–∏–π –≤ –±—É–¥—É—â–µ–º –≤–∞–∂–Ω–æ –≤—ã–¥–µ–ª–∏—Ç—å –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –ø–µ—Ä–∏–æ–¥, –∫–æ—Ç–æ—Ä—ã–π —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ –æ—Ç—Ä–∞–∑–∏—Ç –∞–∫—Ç—É–∞–ª—å–Ω—ã–µ –∏ –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ. –ê–∫—Ç—É–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —É—á–∏—Ç—ã–≤–∞—é—Ç —Ç–µ–∫—É—â–∏–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è –Ω–∞ —Ä—ã–Ω–∫–µ. –ù–æ –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ, –≤ —Å–≤–æ—é –æ—á–µ—Ä–µ–¥—å, —Ç–æ–∂–µ –º–æ–≥—É—Ç –±—ã—Ç—å –∫—Ä–∞–π–Ω–µ –ø–æ–ª–µ–∑–Ω—ã –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è, –≤–æ–∑–º–æ–∂–Ω–æ, –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å –Ω–µ–∫—É—é –º–æ–¥–µ–ª—å –ø–æ–≤–µ–¥–µ–Ω–∏—è, –Ω–æ –≥–ª–∞–≤–Ω–æ–µ ‚Äî –∑–∞—Ç—Ä–∞–≥–∏–≤–∞—Ç—å –∏—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –ø–µ—Ä–∏–æ–¥—ã, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∏–µ—Å—è —Ä–∞–∑–Ω—ã–º–∏ —Ç—Ä–µ–Ω–¥–∞–º–∏ (—á—Ç–æ–±—ã –Ω–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –∫–∞–∫ —Å –º–æ–¥–µ–ª—å—é VaR üòÖ).

### –ò–∑–º–µ–Ω–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–æ–æ–±—â–µ –º–µ–Ω—è–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.
–ü–æ—Å—Ç—Ä–æ–∏–º —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ—Ç—Ä–∞–∑–∏—Ç —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ –∫–∞–∂–¥–æ–π –∏–∑ –ø–ª–∞—Ç—Ñ–æ—Ä–º —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ –≥–æ–¥–∞–º.

–ì—Ä–∞—Ñ–∏–∫ –ø–æ–ª—É—á–∏–ª—Å—è –º–µ–ª–∫–æ–º–∞—Å—à—Ç–∞–±–Ω—ã–º, –Ω–æ —É–∂–µ –ø–æ –Ω–µ–º—É –≤–∏–¥–Ω–æ, —á—Ç–æ —É –ø–ª–∞—Ç—Ñ–æ—Ä–º, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –µ—Å—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–π "—Å—Ä–æ–∫ –∂–∏–∑–Ω–∏", –æ–∫–∞–Ω—á–∏–≤–∞—é—â–∏–π—Å—è —Å–ø–∞–¥–æ–º –ø—Ä–æ–¥–∞–∂. –ü–∏–∫ –ø—Ä–æ–¥–∞–∂ –Ω–∞ –Ω–µ–∫–æ—Ç—Ä—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –ø—Ä—è–º–æ –Ω–∞ –∏—Ö —Å—Ç–∞—Ä—Ç, –Ω–∞ –¥—Ä—É–≥–∏—Ö –∂–µ —Å–ª—É—á–∞–µ—Ç—Å—è –≤ —Å–µ—Ä–µ–¥–∏–Ω–µ –∏—Ö "–∂–∏–∑–Ω–µ–Ω–Ω–æ–≥–æ —Ü–∏–∫–ª–∞". –ù–∞–∏–º–µ–Ω–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–∏ –ª–∏—à—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –æ–¥–Ω–æ–≥–æ –≥–æ–¥–∞!

–¢–µ–ø–µ—Ä—å –≤—ã–±–µ—Ä–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏. –î–ª—è —ç—Ç–æ–≥–æ —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂.

```python
top_platforms = data[data['year_of_release'] != 2049]\
    .groupby(['platform'])\
    .agg({'all_sales':'sum'})\
    .sort_values(by='all_sales', ascending=False)

top_platforms
```

–û—Ç—Å–µ—á—ë–º –ø–µ—Ä–µ—á–µ–Ω—å –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ *XOne* ‚Äî —Å–æ —Å–ª–µ–¥—É—é—â–µ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–æ–π –≤ —Å–ø–∏—Å–∫–µ –±–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –≤ —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂–∞—Ö (–ø–æ—á—Ç–∏ –≤ –¥–≤–∞ —Ä–∞–∑–∞) ‚Äî –∏ —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –∏–∑ –æ—Ç–æ–±—Ä–∞–Ω–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º.

```python
top_platforms = list(top_platforms.head(18).index)
top_platforms
```

–ü–æ—Å—Ç—Ä–æ–∏–º –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –ø–æ –≥–æ–¥–∞–º —É–∂–µ –ª–∏—à—å —Å –≤—ã–±—Ä–∞–Ω–Ω—ã–º–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º–∏.

```python
data_top_platform = data.query('platform == @top_platforms')

sns.scatterplot(x='year_of_release', y='platform',
                hue='platform',
                size='all_sales',
                data=data_top_platform,
                sizes=(30, 700))
plt.xlim(1980, 2017)
plt.title('–ò–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–¥–∞–∂ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏', fontsize=18)
plt.xticks(range(1980,2017), rotation=45)
plt.grid(True, alpha=.3)
plt.legend(bbox_to_anchor=(-0.1, 1));
```

–ù–∞ —Ç–∞–∫–æ–º —É—Å–µ—á—ë–Ω–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ —É–∂–µ –ø—Ä–æ—â–µ –æ—Ü–µ–Ω–∏—Ç—å, –∫–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∫–æ–≥–¥–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–∏ –∏ –∫–∞–∫ –º–µ–Ω—è–ª–∏—Å—å —Å—É–º–º–∞—Ä–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä –Ω–∞ —ç—Ç–∏—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö.

–ß—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, –∑–∞ –∫–∞–∫–æ–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–π —Å—Ä–æ–∫ –ø–æ—è–≤–ª—è—é—Ç—Å—è –Ω–æ–≤—ã–µ –∏ –∏—Å—á–µ–∑–∞—é—Ç —Å—Ç–∞—Ä—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º "–≤—Ä–µ–º—è –∂–∏–∑–Ω–∏" –∫–∞–∂–¥–æ–π –∏–∑ –Ω–∏—Ö.

```python
platform_lt = (data.query('year_of_release!=2049') # –±–µ—Ä—ë–º —Å—Ä–µ–∑ –±–µ–∑ –≥–æ–¥–∞-–∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤
        .loc[:,['year_of_release', 'platform']]    # –∏–∑ —Å—Ä–µ–∑–∞ –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã —Å –ø–ª–∞—Ç—Ñ–æ—Ä–º–æ–π –∏ –≥–æ–¥–æ–º
        .drop_duplicates()                         # –æ—Ç–±—Ä–∞—Å—ã–≤–∞–µ–º –¥—É–±–ª–∏
        # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–µ—Ç —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –≤ —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ:
        .pivot_table(index=['platform'], values='year_of_release', aggfunc='count') 
        .sort_values(by='year_of_release', ascending=False)   # –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é
       )
platform_lt 
```

–î–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –ø–æ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ `platform_lt` –ø–æ—Å—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É, –Ω–∞ –Ω–µ–π –∂–µ –æ—Ç–º–µ—Ç–∏–º –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ‚Äî –æ–Ω–æ –∏ –ø–æ–∫–∞–∂–µ—Ç –Ω–∞–º —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–Ω—ã–π —Å—Ä–æ–∫ "–∂–∏–∑–Ω–∏" –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã.

–î—Ä—É–≥–∞—è —Å–∏—Ç—É–∞—Ü–∏—è –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö _3DS_ –∏ _PC_: –º–µ–¥–∏–∞–Ω–∞ –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–∏ –º–µ–Ω–µ–µ, —á–µ–º 100 —Ç—ã—Å. –∫–æ–ø–∏–π, —Ä–∞–∑–±—Ä–æ—Å –≤ —Ü–µ–ª–æ–º –≥–æ—Ä–∞–∑–¥–æ –º–µ–Ω—å—à–µ, —á—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∏—Ö –∫–∞–∫ –º–µ–Ω–µ–µ –ø—Ä–∏–±—ã–ª—å–Ω—ã–µ, –Ω–æ –±–æ–ª–µ–µ _—Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑—É–µ–º—ã–µ_ –ø–ª–∞—Ñ—Ç—Ñ–æ—Ä–º—ã.

–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞ _PS3_ ‚Äî –Ω–µ—á—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ –º–µ–∂–¥—É —ç—Ç–∏–º–∏ –¥–≤—É–º—è —Å—Ü–µ–Ω–∞—Ä–∏—è–º–∏.

### –í–ª–∏—è–Ω–∏–µ –æ—Ç–∑—ã–≤–æ–≤  –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –≤–ª–∏—è—é—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤–Ω—É—Ç—Ä–∏ –æ–¥–Ω–æ–π –ø–æ–ø—É–ª—è—Ä–Ω–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –æ—Ç–∑—ã–≤—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –ø–æ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ—è–Ω–∏—è.

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –≤ —Ç–µ–ª–µ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç —Å—Ç—Ä–æ–∏—Ç—å—Å—è –¥–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞: –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π –∏–≥—Ä –æ—Ç –æ—Ü–µ–Ω–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –æ—Ü–µ–Ω–æ–∫ –∫—Ä–∏—Ç–∏–∫–æ–≤ –Ω–∞ –∫–∞–∂–¥–æ–π –æ—Ç–¥–µ–ª—å–Ω–æ –≤–∑—è—Ç–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –¢—É—Ç –∂–µ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –∏ –Ω–∞–Ω–µ—Å—ë–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —ç—Ç–∏—Ö –≤–µ–ª–∏—á–∏–Ω.

```python
def score_impact(platf):
    
    # –¥–≤–µ —Ç–∞–±–ª–∏—Ü—ã, —á—Ç–æ–±—ã —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –º–∞–∫—Å–∏–º—É–º –¥–∞–Ω–Ω—ã—Ö: –≤ —Å—Ç—Ä–æ–∫–µ –º–æ–∂–µ—Ç –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω–∞ –∏–∑ –æ—Ü–µ–Ω–æ–∫:
    # –∏–∑–±–∞–≤–ª—è–µ–º—Å—è –æ—Ç –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–æ–ø—É—Å–∫–æ–≤ —Ä–µ–π—Ç–∏–Ω–≥–∞—Ö (-1)
    user_score_data = actual_data.query('user_score >= 0 and platform == @platf').reset_index(drop=True)
    critic_score_data = actual_data.query('critic_score >= 0  and platform == @platf').reset_index(drop=True)
    
    fig, axis = plt.subplots(1,2)
    
    # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Ç–µ–∫—Å—Ç–æ–≤–æ–π –ø–ª–∞—à–∫–∏
    props = dict(boxstyle='round', facecolor='thistle', alpha=0.5)
    
    # –¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è –¥–ª—è –æ—Ü–µ–Ω–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
    sns.scatterplot(
        data=user_score_data,
        x='user_score',
        y='all_sales',
        hue = 'user_score',
        size = 'user_score',
        sizes = (10,100), 
        ax=axis[0]
    ).set(title='–æ—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Äî –ø—Ä–æ–¥–∞–∂–∏')
    
    # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ '–æ—Ü–µ–Ω–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Äî –ø—Ä–æ–¥–∞–∂–∏'
    text_1 = f'–ö–æ—ç—Ñ.–∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: {user_score_data["all_sales"].corr(user_score_data["user_score"]):.2f}'
    plt.text(0.02, 0.5, text_1, 
             horizontalalignment='center', verticalalignment='center', 
             transform=ax.transAxes, bbox=props, fontsize='large')

    # –¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è –¥–ª—è –æ—Ü–µ–Ω–æ–∫ –∫—Ä–∏—Ç–∏–∫–æ–≤
    sns.scatterplot(
        data=critic_score_data,
        x='critic_score',
        y='all_sales',
        hue = 'critic_score',
        size = 'critic_score',
        sizes = (10,100), 
        ax=axis[1]
    ).set(title='–æ—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤ ‚Äî –ø—Ä–æ–¥–∞–∂–∏')
    
    # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ '–æ—Ü–µ–Ω–∫–∞ –∫—Ä–∏—Ç–∏–∫–æ–≤ ‚Äî –ø—Ä–æ–¥–∞–∂–∏'
    text_2 = f'–ö–æ—ç—Ñ.–∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: {critic_score_data["all_sales"].corr(critic_score_data["critic_score"]):.2f}'
    plt.text(0.55, 0.5, text_2, 
             horizontalalignment='center', verticalalignment='baseline', 
             transform=ax.transAxes, bbox=props, fontsize='large')
    
    # –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –≥—Ä–∞–Ω–∏—Ü—ã –≥—Ä–∞—Ñ–∏–∫–æ–≤ –¥–ª—è –æ—Å–∏ —Ö
    axis[0].set_xlim(0,10)
    axis[1].set_xlim(0,100)
    
    fig.suptitle(f'–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞ {platf}', fontsize=18)
```

–ò –ø—Ä–∏–º–µ–Ω–∏–º –Ω–∞–ø–∏—Å–∞–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é –≤ —Ü–∏–∫–ª–µ –∫ –∫–∞–∂–¥–æ–π –∏–∑ —à–µ—Å—Ç–∏ –æ—Ç–æ–±—Ä–∞–Ω–Ω—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º:

```python
for i in top_6_platforms:
    
    score_impact(i)
```

–ü–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω—ã–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—è–º –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Ç–æ–º, —á—Ç–æ –Ω–∏ –≤ –æ–¥–Ω–æ–º –∏–∑ —à–µ—Å—Ç–∏ —Å–ª—É—á–∞–µ–≤ _–æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–µ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç —Å –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä_. –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å–∫–ª–æ–Ω–Ω—ã –æ—Ü–µ–Ω–∏–≤–∞—Ç—å –∏–≥—Ä—ã –≤ 5‚Äî8 –±–∞–ª–ª–æ–≤, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —á–∞—Å—Ç–æ –≤—ã—Å–æ–∫–æ –æ—Ü–µ–Ω–∏–≤–∞—é—Ç —É—Å–ø–µ—à–Ω—ã–µ —Ä–µ–ª–∏–∑—ã, –æ–¥–Ω–∞–∫–æ —á—ë—Ç–∫–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ç—É—Ç –Ω–µ –≤—ã—Ä–∏—Å–æ–≤—ã–≤–∞–µ—Ç—Å—è.

–°—Ä–∞–≤–Ω–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª–µ–µ –∑–∞–º–µ—Ç–Ω–∞—è, –Ω–æ –≤—Å—ë –∂–µ –¥–æ–≤–æ–ª—å–Ω–æ _—Å–ª–∞–±–∞—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤_. –ù–∞–∏–±–æ–ª–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∞ –æ–Ω–∞ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö _XOne_ –∏ _PS4_, –Ω–∞–∏–º–µ–Ω–µ–µ (–ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç) ‚Äî –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ _PC_.

–¢–∞–∫–∂–µ —Å—Ç–æ–∏—Ç –æ—Ç–º–µ—Ç–∏—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—É—é —Å–æ–ª–∏–¥–∞—Ä–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏ –∫—Ä–∏—Ç–∏–∫–∞–º–∏ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ _3DS_, —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω—ë–Ω–Ω–æ–π –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –≤ —è–ø–æ–Ω—Å–∫–æ–º —Ä–µ–≥–∏–æ–Ω–µ.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º–æ–∂–Ω–æ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –º–æ–∂–Ω–æ (–Ω–æ –æ—Å—Ç–æ—Ä–æ–∂–Ω–æ) –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è —É—Å–ø–µ—Ö–∞ –≤—ã–ø—É—Å–∫–∞–µ–º—ã—Ö –∏–≥—Ä, –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–ª—è —ç—Ç–∏—Ö —Ü–µ–ª–µ–π –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –±–µ—Å–ø–æ–ª–µ–∑–Ω—ã.

### –ü—Ä–∏–±—ã–ª—å–Ω—ã–µ –∂–∞–Ω—Ä—ã

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –æ–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–≥—Ä –ø–æ –∂–∞–Ω—Ä–∞–º. –°–Ω–∞—á–∞–ª–∞ —Å—Ä–∞–≤–Ω–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π –ø–æ –∂–∞–Ω—Ä–∞–º –∑–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥:

–ê —Ç–µ–ø–µ—Ä—å –ø–æ—Å–º–æ—Ç—Ä–∏–º —Ç–æ –∂–µ –≤ –¥–∏–Ω–∞–º–∏–∫–µ –ø–æ –≥–æ–¥–∞–º:

–ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –æ–±—â–∏–π **–Ω–∏—Å—Ö–æ–¥—è—â–∏–π —Ç—Ä–µ–Ω–¥ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –∏–≥—Ä** –≤ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 4 –≥–æ–¥–∞ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –∂–∞–Ω—Ä–∞, **—Ç–æ–ø–æ–≤—ã–º –∂–∞–Ω—Ä–æ–º** –Ω–∞ –ø—Ä–æ—Ç—è–∂–µ–Ω–∏–∏ —ç—Ç–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —É–≤–µ—Ä–µ–Ω–Ω–æ –æ—Å—Ç–∞—ë—Ç—Å—è _Shooter_. –î–æ–≥–æ–Ω—è—Ç—å –µ–≥–æ –ø—ã—Ç–∞—é—Ç—Å—è –∏–≥—Ä—ã –≤ –∂–∞–Ω—Ä–µ _Sports_.

–°—Ç—Ä–µ–º–∏—Ç–µ–ª—å–Ω—ã–π —Ä–æ—Å—Ç –∏–Ω–¥—É—Å—Ç—Ä–∏–∏ —Å —Å–µ—Ä–µ–¥–∏–Ω—ã 90-—Ö –¥–æ—Å—Ç–∏–≥ –ø–∏–∫–∞ –≤ 2008-–æ–º, –ø–æ—Å–ª–µ —á–µ–≥–æ –Ω–∞—á–∞–ª–æ—Å—å **–µ–∂–µ–≥–æ–¥–Ω–æ–µ –ø–∞–¥–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã–ø—É—Å–∫–∞–µ–º—ã—Ö –∏–≥—Ä**. –ü—Ä–∏—á–∏–Ω–∞ –º–æ–∂–µ—Ç –∫—Ä—ã—Ç—å—Å—è –≤ –Ω–∞—Ä–∞—Å—Ç–∞–Ω–∏–∏ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –º–æ–±–∏–ª—å–Ω—ã—Ö –∏–≥—Ä –∏–ª–∏ –≤ –ø–µ—Ä–µ—Ö–æ–¥–µ –Ω–∞ –±–æ–ª–µ–µ –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—É—é, "–≥–ª—É–±–æ–∫—É—é" —Ä–∞–∑—Ä–∞–±–æ—Ç–∫—É –≤–∏–¥–µ–æ-–∏–≥—Ä, —á—Ç–æ –Ω–µ–∏–∑–±–µ–∂–Ω–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç —Å—Ä–æ–∫ –∏—Ö –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤–∞ –∏ —Å–æ–∫—Ä–∞—â–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—Å–∫–∞.

–ò–≥—Ä–æ–≤—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –∏–º–µ—é—Ç –Ω–µ–∫–æ—Ç–æ—Ä—ã–π "—Å—Ä–æ–∫ –∂–∏–∑–Ω–∏": –Ω–∞–∏–±–æ–ª–µ–µ —É—Å–ø–µ—à–Ω—ã –æ–Ω–∏ –±—ã–≤–∞—é—Ç –≤ —Å–∞–º–æ–º –Ω–∞—á–∞–ª–µ –∏–ª–∏ —Å–µ—Ä–µ–¥–∏–Ω–µ —Å–≤–æ–µ–≥–æ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è, –∏ –∑–∞—Ç–µ–º –∑–∞ –Ω–µ–∏–∑–±–µ–∂–Ω—ã–º —Å–ø–∞–¥–æ–º –ø—Ä–æ–¥–∞–∂ —Å–ª–µ–¥—É–µ—Ç –ø—Ä–µ–∫—Ä–∞—â–µ–Ω–∏–µ –∏—Ö —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è. **–•–∞—Ä–∞–∫—Ç–µ—Ä–Ω–∞—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ‚Äî –æ–∫–æ–ª–æ 7 –ª–µ—Ç**.  –ù–∞–∏–º–µ–Ω–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª–∏ –ª–∏—à—å –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –æ–¥–Ω–æ–≥–æ –≥–æ–¥–∞.

–î–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ —Ç–µ—Ö –∏–ª–∏ –∏–Ω—ã—Ö –∏–≥—Ä –≤ –±—É–¥—É—â–µ–º –æ–ø—Ä–µ–¥–µ–ª—ë–Ω **–∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥ ‚Äî 4 –≥–æ–¥–∞**. –≠—Ç–æ —á—É—Ç—å –±–æ–ª—å—à–µ –ø–æ–ª–æ–≤–∏–Ω—ã –∂–∏–∑–Ω–µ–Ω–Ω–æ–≥–æ —Ü–∏–∫–ª–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º, —Å —É—á—ë—Ç–æ–º –≤–æ–∑–º–æ–∂–Ω–æ–π –Ω–µ–ø–æ–ª–Ω–æ—Ç—ã –¥–∞–Ω–Ω—ã—Ö –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–π, 2016-—ã–π –≥–æ–¥.

–ì–ª–æ–±–∞–ª—å–Ω—ã–µ **–ø—Ä–æ–¥–∞–∂–∏ –ø–∞–¥–∞—é—Ç** –∞–±—Å–æ–ª—é—Ç–Ω–æ –Ω–∞ –≤—Å–µ—Ö –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –≤ —Ç–µ—á–µ–Ω–∏–µ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —á–µ—Ç—ã—Ä—ë—Ö –ª–µ—Ç. –í—ã–¥–µ–ª–µ–Ω—ã **–ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –ø—Ä–∏–±—ã–ª—å–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã**: _XOne, PS4, 3DS, X360, PS3 –∏ PC_ ‚Äî —ç—Ç–æ —Å–∞–º—ã–µ –ø—Ä–∏–±—ã–ª—å–Ω—ã–µ –∏ —Å–∞–º—ã–µ —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã. 

–ú–µ–∂–¥—É **–æ—Ü–µ–Ω–∫–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏–≥—Ä –Ω–µ—Ç –≤—ã—Ä–∞–∂–µ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏**, –∞ –≤–æ—Ç **–æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤ –∏–º–µ—é—Ç —É–º–µ—Ä–µ–Ω–Ω—É—é –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—É—é –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é** —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π –∏–≥—Ä ‚Äî –∏—Ö –º–æ–∂–Ω–æ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –≤–æ –≤–Ω–∏–º–∞–Ω–∏–µ –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –ø—Ä–æ–≥–Ω–æ–∑–∞.

**–ù–∞–∏–±–æ–ª–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–º –∂–∞–Ω—Ä–æ–º** –Ω–∞ –ø—Ä–æ—Ç—è–∂–µ–Ω–∏–∏ –≤—Å–µ–≥–æ –∞–∫—Ç—É–∞–ª—å–Ω–æ–≥–æ –ø–µ—Ä–∏–æ–¥–∞ –æ—Å—Ç–∞–≤–∞–ª—Å—è _Shooter_, **–Ω–∞–∏–º–µ–Ω–µ–µ –≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω—ã** ‚Äî _Strategy_, _Adventure_ –∏ _Puzzle_.

## –°–æ—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –ø–æ—Ä—Ç—Ä–µ—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

–û–ø—Ä–µ–¥–µ–ª–∏–º –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞ (NA, EU, JP).

### –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

–í—ã—è–≤–∏–º —Ç–æ–ø-5 –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –≤ –∫–∞–∂–¥–æ–º –∏–∑ —Ä–µ–≥–∏–æ–Ω–æ–≤. –°—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –±—É–¥–µ–º –¥–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –Ω–∞ –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ.

```python
def top_region_share(index, reg, name_reg, bar_color): 
    
    # —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤:
            # –∏–Ω–¥–µ–∫—Å –¥–ª—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏, 
            # —Ä–µ–≥–∏–æ–Ω –ø—Ä–æ–¥–∞–∂, 
            # –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–∞ –¥–ª—è –ø–æ–¥–ø–∏—Å–∏ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ 
            # —Ü–≤–µ—Ç –¥–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–∏–∞–≥—Ä–∞–º–º—ã;
            
    # —Ñ–æ—Ä–º–∏—Ä—É–µ—Ç —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –∏–Ω–¥–µ–∫—Å—É –∏ —Ä–µ–≥–∏–æ–Ω—É –∑–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã–π –ø–µ—Ä–∏–æ–¥:
       
    region_data = data.query('2013 <= year_of_release <= 2016')\
    .pivot_table(index=index, values=reg, aggfunc='sum')
    
    # –¥–æ–±–∞–≤–ª—è–µ—Ç –≤ —Ç–∞–±–ª–∏—Ü—É —Å—Ç–æ–ª–±–µ—Ü —Å –¥–æ–ª–µ–π –ø—Ä–æ–¥–∞–∂ –¥–ª—è –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –≤ —Ä–µ–≥–∏–æ–Ω–µ 
    # –æ—Ç —Å—É–º–º–∞—Ä–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –≤ —Ä–µ–≥–∏–æ–Ω–µ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö:
    
    region_data[reg+'_share'] = region_data[reg] / region_data[reg].sum() * 100
    
    # —Å–æ—Ä—Ç–∏—Ä—É–µ—Ç —Ç–∞–±–ª–∏—Ü—É –∏ —Å—Ç—Ä–æ–∏—Ç —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ç–æ–ø-5 –ø–æ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–º—É –≤ —Ñ—É–Ω–∫—Ü–∏—é –∏–Ω–¥–µ–∫—Å—É –≤ —Ä–µ–≥–∏–æ–Ω–µ:
    
    region_data\
    .sort_values(by=reg)\
    .tail()\
    .plot(
        kind='barh', 
        y=reg+'_share',
        color=bar_color,
        legend=True,
        figsize=(14,4)
    )
    plt.title(f'–¢–æ–ø-5 –≤ {name_reg}', fontsize=18)
    plt.xlim(0,50) # –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π –º–∞—Å—à—Ç–∞–± –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–æ–≤: –æ—Ç 0 –¥–æ 50 %
    plt.xlabel('sales_share in %')
```

```python
regions = {'na_sales': 'NA', 'eu_sales': 'EU', 'jp_sales': 'JP'} # —Å–ª–æ–≤–∞—Ä—å —Å —Ä–µ–≥–∏–æ–Ω–∞–º–∏

# –∏ –ø—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ –∫–∞–∂–¥–æ–º—É —Ä–µ–≥–∏–æ–Ω—É –ø–æ –æ—á–µ—Ä–µ–¥–∏. –†–µ–≥–∏–æ–Ω—ã –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –∏–∑ —Å–ª–æ–≤–∞—Ä—è:

for key, value in regions.items():
    top_region_share('platform', key, value, 'burlywood')
```

–í –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ (NA) –∏ –ï–≤—Ä–æ–ø–µ (EU) —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ö–æ–∂–µ–µ: –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ _PS4_ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —Å –∑–∞–º–µ—Ç–Ω—ã–º –æ—Ç—Ä—ã–≤–æ–º: –≤ NA –Ω–∞ –µ—ë –¥–æ–ª—é –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –æ–∫–æ–ª–æ 25 % –ø—Ä–æ–¥–∞–∂, –∞ –≤ EU –≤—Å–µ 36 %. –¢–æ–ø –≤ —ç—Ç–∏—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –æ–¥–Ω–∏ –∏ —Ç–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ‚Äî _PS3, XOne, X360, 3DS_ , ‚Äî –ª–∏—à—å –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –≤–∞—Ä—å–∏—Ä—É—è—Å—å –ø–æ –º–µ—Å—Ç–∞–º. –°—Ä–µ–¥–∏ –µ–≤—Ä–æ–ø–µ–π—Å–∫–∏—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ _PS3_ –≥–æ—Ä–∞–∑–¥–æ –ø–æ–ø—É–ª—è—Ä–Ω–µ–µ, —á–µ–º —É —Å–µ–≤–µ—Ä–æ–∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∏—Ö, –∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ _3DS_ –∏ —Ç–∞–º –∏ —Ç–∞–º —Å–ø–∏—Å–æ–∫ –∑–∞–º—ã–∫–∞–µ—Ç (–æ–∫–æ–ª–æ 8 % –ø—Ä–æ–¥–∞–∂).

–°–æ–≤—Å–µ–º –¥—Ä—É–≥–∞—è —Å–∏—Ç—É–∞—Ü–∏—è –≤ –Ø–ø–æ–Ω–∏–∏ (JP): —Ç—É—Ç —Å –¥—Ä–∞–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ—Ç—Ä—ã–≤–æ–º –≤–ø–µ—Ä—ë–¥ –≤—ã—Ä—ã–≤–∞–µ—Ç—Å—è –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞ _3DS_ (–ø–æ—á—Ç–∏ 50 %!). –û—Å—Ç–∞–ª—å–Ω—ã–µ –≤—Ö–æ–¥—è—â–∏–µ –≤ —Å–ø–∏—Å–æ–∫ –Ω–∞–∏–±–æ–ª–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö ‚Äî _PS3, PSV, PS4 –∏ WiiU_ ‚Äî –æ—Ç—Å—Ç–∞—é—Ç –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ ‚Äî –ø—Ä–æ–¥–ø–∂–∏ –æ—Ç 8 –¥–æ 17 %. –í—Ö–æ–¥—è—â–∏–µ –∑–¥–µ—Å—å –≤ —Å–ø–∏—Å–æ–∫ –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö _PSV –∏ WiiU_ –≤ NA –∏ EU –Ω–µ —Ñ–∏–≥—É—Ä–∏—Ä—É—é—Ç –≤–æ–≤—Å–µ.

### –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã


–ê–Ω–∞–ª–æ–≥–∏—á–Ω—ã–º –æ–±—Ä–∞–∑–æ–º —Å–æ—Å—Ç–∞–≤–∏–º —Ç–æ–ø-5 –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ –ø–æ –∂–∞–Ω—Ä–∞–º. –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ç—É –∂–µ —Ñ—É–Ω–∫—Ü–∏—é `top_region_share`.

```python
for key, value in regions.items():
    top_region_share('genre', key, value, 'plum')
```

–°–Ω–æ–≤–∞ –≤–∏–¥–∏–º, –∫–∞–∫ –ø–æ—Ö–æ–∂–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≤ NA –∏ EU: —Ç–æ–ø –∂–∞–Ω—Ä–æ–≤, –∏—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ –¥–æ–ª–∏ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –∏–¥–µ–Ω—Ç–∏—á–Ω—ã. –û–∫–æ–ª–æ 30 % –ø—Ä–æ–¥–∞–∂ –≤ —ç—Ç–∏—Ö —Ä–µ–≥–∏–æ–Ω–∞—Ö –ø—Ä–∏—Ö–æ–¥—è—Ç—Å—è –Ω–∞ –∏–≥—Ä—ã –≤ –∂–∞–Ω—Ä–µ _Action_. –î–∞–ª–µ–µ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ —Å–ª–µ–¥—É—é—Ç _Shooter, Sports_ –∏ _Role-playing_. –ó–∞–º—ã–∫–∞—é—Ç —Ç–æ–ø—ã –∂–∞–Ω—Ä—ã _Misc_ ‚Äî –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ, ‚Äî –∏ _Racing_ –≤ –ï–≤—Ä–æ–ø–µ.

–í –Ø–ø–æ–Ω–∏–∏ –±–æ–ª–µ–µ 35 % –ø—Ä–æ–¥–∞–∂ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –∏–≥—Ä—ã –≤ –∂–∞–Ω—Ä–µ _Role-playing_, –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –æ—Ç—Å—Ç–∞—ë—Ç –∂–∞–Ω—Ä _Action_ (–ø–æ—á—Ç–∏ 30 %), –∏ –µ—â—ë –ø—Ä–∏–º–µ—Ä–Ω–æ 15 % –ø—Ä–æ–¥–∞–∂ —Å—É–º–º–∞—Ä–Ω–æ –¥–µ–ª—è—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –ø–æ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç–∏ _Misc, Fighting_ –∏ _Shooter_.

### –í–ª–∏—è–Ω–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∞ ESRB –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏

**–ù0**: —Å—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã _Xbox One_ ***—Ä–∞–≤–µ–Ω*** —Å—Ä–µ–¥–Ω–µ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º—É —Ä–µ–π—Ç–∏–Ω–≥—É –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã _PC_.

**–ù1**: —Å—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã _Xbox One_ ***–æ—Ç–ª–∏—á–∞–µ—Ç—Å—è*** –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã _PC_.

–°—Ä–∞–≤–Ω–∏–≤–∞–µ–º—ã–µ –≤—ã–±–æ—Ä–∫–∏ *–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã* –º–µ–∂–¥—É —Å–æ–±–æ–π, –ø–æ—ç—Ç–æ–º—É –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å *t-–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞ –¥–ª—è –Ω–µ—Å–≤—è–∑–∞–Ω–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫*. –° –µ–≥–æ –ø–æ–º–æ—â—å—é –º–æ–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑–ª–∏—á–∞—é—Ç—Å—è –∏—Ö —Å—Ä–µ–¥–Ω–∏–µ.
–¢.–∫. —Ñ–æ—Ä–º—É–ª–∞ t-–∫—Ä–∏—Ç–µ—Ä–∏—è –≤–∫–ª—é—á–∞–µ—Ç –≤ —Å–µ–±—è —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫, —Ç–æ –∫—Ä–∏—Ç–µ—Ä–∏–π –º–æ–∂–µ—Ç –¥–∞–≤–∞—Ç—å –Ω–µ–∞–¥–µ–∫–≤–∞—Ç–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø—Ä–∏ –Ω–∞–ª–∏—á–∏–∏ –≤—ã–±—Ä–æ—Å–æ–≤. –í —Ä–∞–∑–¥–µ–ª–µ –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –º—ã —É–∂–µ —É–±–µ–¥–∏–ª–∏—Å—å, —á—Ç–æ –≤ –∫–æ–ª–æ–Ω–∫–µ –Ω–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π-–≤—ã–±—Ä–æ—Å–æ–≤, –æ–¥–Ω–∞–∫–æ –ø–µ—Ä–µ–¥ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ–º –≤—ã–±–æ—Ä–æ–∫ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –æ—Ç—Å–µ—è—Ç—å –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ (= -1) –≤ —Å—Ç–æ–ª–±—Ü–µ `user_score`.

–°—Ñ–æ—Ä–º–∏—Ä—É–µ–º –º–∞—Å—Å–∏–≤—ã `data_xone` –∏ `data_pc` —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤ –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∏—Ö –≤ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–∞–ª–µ–µ.

```python
data_xone = data.query('2013 <= year_of_release <= 2016 and platform == "XOne" and user_score >= 0')['user_score']
data_pc = data.query('2013 <= year_of_release <= 2016 and platform == "PC" and user_score >= 0')['user_score']
```

–ß—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `equal_var` —Ñ—É–Ω–∫—Ü–∏–∏ `st.ttest_ind`, —Å—Ä–∞–≤–Ω–∏–º –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫. –î–∏—Å–ø–µ—Ä—Å–∏–∏ –≤ –Ω–∏—Ö –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –Ω–µ –º–æ–≥—É—Ç, —Ç–∞–∫ –∫–∞–∫ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–µ–π—Ç–∏–Ω–≥–∞ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω—ã –¥–∏–∞–ø–∞–∑–æ–Ω–æ–º –æ—Ç 0 –¥–æ 10 –±–∞–ª–ª–æ–≤.

```python
display(len(data_xone), len(data_pc))
```

–†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è, –Ω–æ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ ‚Äî –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `equal_var` –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é.

–ò –º–æ–∂–Ω–æ –Ω–∞–∫–æ–Ω–µ—Ü –ø—Ä–æ–≤–µ—Ä—è—Ç—å *–≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö* –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π –ø–æ –≤–∑—è—Ç—ã–º –∏–∑ –Ω–∏—Ö –≤—ã–±–æ—Ä–∫–∞–º. –ö—Ä–∏—Ç–µ—Ä–∏–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ `alpha` –∑–∞–¥–∞–¥–∏–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π ‚Äî 5 %.

```python
alpha = 0.05

results_1 = st.ttest_ind(data_xone, data_pc)
display(f'–∑–Ω–∞—á–µ–Ω–∏–µ p-value: {results_1.pvalue}')
results_1.pvalue < alpha
```
